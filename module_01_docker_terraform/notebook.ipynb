```python
import pandas as pd

#data URLs
green_url = "https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet"
zones_url = "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv"

# Read data
df_green = pd.read_parquet(green_url)
df_zones = pd.read_csv(zones_url)

df_green.head(), df_zones.head()
```

```python
from sqlalchemy import create_engine

engine = create_engine(
    "postgresql://root:root@localhost:5432/ny_taxi"
)

engine
```

```python
df_green.head(0).to_sql(
    name="green_tripdata",
    con=engine,
    if_exists="replace",
    index=False
)

df_zones.head(0).to_sql(
    name="zones",
    con=engine,
    if_exists="replace",
    index=False
)
```

```python
df_green.to_sql(
    name="green_tripdata",
    con=engine,
    if_exists="append",
    index=False
)

df_zones.to_sql(
    name="zones",
    con=engine,
    if_exists="append",
    index=False
)
```
